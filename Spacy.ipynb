## Paradigma do SpaCy



```python
import spacy
from spacy import displacy
nlp = spacy.load("pt_core_news_sm")
doc = nlp("Eu amo a srie Star Wars,  um dos meus filmes favoritos, embora eu tambm goste de Lord of The Rings")
```

```python
doc[0] #Token
```

```python
doc[0:6] #Span
```

```python
# Dependncia com POS
displacy.render(doc, style="dep")
```

```python
displacy.render(doc, style="ent")
```

## O que realmente acontece quando chamamos `nlp("...")`?

Toda a string  passada por meio de uma pipeline!



* Todas as etapas so customizveis, o que  uma grande vantagem ao utilizar o SpaCy, pois fornece uma API muito otimizada que  altamente modular.
