## Paradigma do SpaCy



```python
import spacy
from spacy import displacy
nlp = spacy.load("pt_core_news_sm")
doc = nlp("Eu amo a série Star Wars, é um dos meus filmes favoritos, embora eu também goste de Lord of The Rings")
```

```python
doc[0] #Token
```

```python
doc[0:6] #Span
```

```python
# Dependência com POS
displacy.render(doc, style="dep")
```

```python
displacy.render(doc, style="ent")
```

## O que realmente acontece quando chamamos `nlp("...")`?

Toda a string é passada por meio de uma pipeline!



* Todas as etapas são customizáveis, o que é uma grande vantagem ao utilizar o SpaCy, pois fornece uma API muito otimizada que é altamente modular.
